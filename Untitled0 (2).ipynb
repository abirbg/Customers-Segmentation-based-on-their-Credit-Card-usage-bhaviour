```python
columns=['PURCHASES', 'ONEOFF_PURCHASES', 'CASH_ADVANCE'
        ]
data=df
for c in columns:
    data.loc[((data[c]>2000)),c]=2000
columns=['BALANCE', 'INSTALLMENTS_PURCHASES', 
        'PAYMENTS', 'MINIMUM_PAYMENTS']
for c in columns:
    
    data.loc[((data[c]>5000)),c]=5000  
data.loc[((data['MINIMUM_PAYMENTS']>2000)),'MINIMUM_PAYMENTS']=2000
data.loc[((data['INSTALLMENTS_PURCHASES']>2000)),'INSTALLMENTS_PURCHASES']=2000
data.loc[((data['BALANCE_FREQUENCY']<0.8)),'BALANCE_FREQUENCY']=0.8
data.loc[((data['ONEOFF_PURCHASES_FREQUENCY']>0.4)),'ONEOFF_PURCHASES_FREQUENCY']=0.4
data.loc[((data['CASH_ADVANCE_TRX']>10)),'CASH_ADVANCE_TRX']=10
data.loc[((data['PURCHASES_TRX']>50)),'PURCHASES_TRX']=50
data.loc[((data['PRC_FULL_PAYMENT']>0.2)),'PRC_FULL_PAYMENT']=0.2
data.loc[((data['TENURE']<11.5)),'TENURE']=11.5
data.loc[((data['TENURE']>12.5)),'TENURE']=12.5
data.loc[((data['CREDIT_LIMIT']>15000)),'CREDIT_LIMIT']=15000
data.loc[((data['CASH_ADVANCE_FREQUENCY']>0.5)),'CASH_ADVANCE_FREQUENCY']=0.5      
```

```python
plt.figure(figsize=(20,35))
for i, col in enumerate(data.columns):
    if data[col].dtype != 'object':
        ax = plt.subplot(9, 2, i+1)
        sns.kdeplot(data[col], ax=ax)
        plt.xlabel(col)
        
plt.show()
```

```python

data.drop(['CUST_ID' ], axis=1, inplace=True)
X= np.asarray(data)
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = scale.fit_transform(X)
X.shape
```

```python
from sklearn.cluster import KMeans

kmeans_models = [KMeans(n_clusters=k, random_state=23).fit(X) for k in range (1, 20)]
innertia = [model.inertia_ for model in kmeans_models]

plt.plot(range(1, 20), innertia)
plt.title('Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

You see the elbow there? It seems like the elbow is around  6 or 7 . 

```python
kmean= KMeans(6)
kmean.fit(X)
labels=kmean.labels_
```

```python
clusters=pd.concat([data, pd.DataFrame({'cluster':labels})], axis=1)
clusters.head(10)
```

```python
for c in clusters:
    grid= sns.FacetGrid(clusters, col='cluster')
    grid.map(plt.hist, c)
```

Visualization of Clusters
Using PCA to transform data to 2 dimensions for visualization

```python
from sklearn.decomposition import PCA
pca = PCA(2)
pca.fit(X)
```

```python
X_PCA = pca.transform(X)
X_PCA.shape
x, y = X_PCA[:, 0], X_PCA[:, 1]

colors = {0: 'red',
          1: 'blue',
          2: 'green', 
          3: 'yellow', 
          4: 'pink',  
          5:'black'}

names = {0: 'who make all type of purchases', 
         1: 'more people with due payments', 
         2: 'who purchases mostly in installments', 
         3: 'who take more cash in advance', 
         4: 'who make expensive purchases',
         5:'who don\'t spend much money'}
  
df = pd.DataFrame({'x': x, 'y':y, 'label':labels}) 
groups = df.groupby('label')

fig, ax = plt.subplots(figsize=(20, 13)) 

for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5,
            color=colors[name],label=names[name], mec='none')
    ax.set_aspect('auto')
    ax.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off')
    ax.tick_params(axis= 'y',which='both',left='off',top='off',labelleft='off')
    
ax.legend()
ax.set_title("Customers Segmentation based on their Credit Card usage bhaviour.")
plt.show()
```
